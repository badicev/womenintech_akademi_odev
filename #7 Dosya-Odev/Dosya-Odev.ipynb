```python
def metin_temizle(filename):
#Dosyayı oku.
    import re
    file = open(filename, 'rt', encoding='utf-8')
    text = file.read()
    file.close()

    
    text = text.lower()
     #Türkçe karakterleri ve internet sayfasında satır atlamalarını kaldırmak için
    turkce_yabancı_sozluk = {'ç': 'c', 'ğ': 'g', 'ı': 'i', 'ö': 'o', 'ş': 's',
    'ü': 'u', "▼": "", "\n": "", "'": ""}

    for key, value in turkce_yabancı_sozluk.items():
        text = text.replace(key, value)

    
   
    #regex ile 3 ile daha fazla sayıdan oluşan kelimeleri kaldır. Yabancı karakterleri kaldır. Boşlukları kaldır.
    patterns = ('/(\S)\1(\1)+/g',  '/\d{3,}/g') #r'[\n\r]') 
    for pattern in patterns:
        text = re.sub(pattern, '', text)
     #   text = re.sub(r'[^\w\s]','',text)        
    
    
    #Harfleri küçült.
    text = [word.lower() for word in text.split()] 

    
            
    

    #Noktalama işaretlerini kaldır.
    table = str.maketrans('', '', ("!#$%&'()*+,-/:;.<=>?@[\]^_`{|}~")) 
    text = [w.translate(table) for w in text]

    
   
    #sayilar_listesi = re.findall(r'\d+', text) ???

        #Temizlenmiş dosyayı yazdır.
        #ÖNEMLİ NOT: x kullandın, kodu yeniden çalıştırmadan önce oluşan dosyayı kapatman gerekiyor.
    with open("temiz_metin.txt", "x", encoding = "utf-8") as f:
        for word in text:
            f.write(word + " ")


    

    
        #Sayıları yeni bir klasöre aktarmak için:
    with open("sayilar.txt", "w", encoding = "utf-8") as f:
        for word in text:
            res = re.findall(r'(\w+?)(\d+)', word)
            if res:
                with open('sayilar.txt', 'a', encoding = "utf-8") as f:
                    for i in res:
                        f.write(i[1] + "\n ")

    
      


                        
            
                

 
  


    #Kelime sayısını takip etmek için:
    word_count = {}

    for word in text:
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1

   
    #En sık kullanılan 10 kelimeyi bulmak için: 
    sorted_word_count = sorted(word_count.items(), key=lambda x: x[1], reverse=True)
        
    print(sorted_word_count[:100])
     #en uzun kelimeyi bulmak için:
    uzunluk = 0
    for word in sorted_word_count:
        if len(word[0]) > uzunluk:
            uzunluk = len(word[0])
            uzun_kelime = word[0]
             
           
    
    with open("en_cok_kullanılan_100_kelime.txt", "w", encoding = "utf-8") as f:
           for word in sorted_word_count:
                f.write(word[0] + (len(uzun_kelime)//4 - len(word[0])) * " " )
                for i in word[0]:
                    f.write(str(i).upper() + " ")
                f.write("\n")
              

               

                




    return 

  

     


metin_temizle('all_data.txt')
```
